```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("C:\\Users\\svish\\OneDrive\\Desktop\\titanic-dataset\\train.csv")

# Drop missing values and select features
df = df[['Age', 'Fare']].dropna()

X = df[['Age']]  # input feature
y = df['Fare']   # target

# Create and train model
model = LinearRegression()
model.fit(X, y)

# Predict fare for age = 22
fare_for_22 = model.predict([[22]])
print(f"Predicted fare for age 22: ₹{fare_for_22[0]:.2f}")

# Predict fares for all data points to draw line
predictions = model.predict(X)

# Plot
plt.scatter(df['Age'], df['Fare'], color='blue', label='Actual Fares')
plt.plot(df['Age'], predictions, color='red', label='Regression Line')
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Linear Regression - Age vs Fare")
plt.legend()
plt.show()
```
