### Matplotlib practice 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(1, 6)
y = np.array([10, 20, 15, 25, 30])
z = np.array([30, 25, 20, 15, 10])
```

1. Line Plot

```python
plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

2. Scatter Plot


```python
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()
```

3. Bar Plot

```python
plt.bar(x, y)
plt.title("Bar Plot")
plt.show()
```

4. Horizontal Bar Plot

```python
plt.barh(x, y)
plt.title("Horizontal Bar Plot")
plt.show()
```

5. Histogram

```python
data = [12, 15, 20, 22, 23, 24, 25, 28, 30, 33, 35, 40]
plt.hist(data, bins=5)
plt.title("Histogram")
plt.show()
```

6. Pie Chart


```python
labels = ['Python', 'C++', 'Java', 'Ruby']
sizes = [215, 130, 245, 210]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

7. Add Grid

```python
plt.plot(x, y)
plt.grid(True)
plt.title("Grid Example")
plt.show()
```

8. Multiple Lines in One Plot

```python
plt.plot(x, y, label='Line 1')
plt.plot(x, z, label='Line 2')
plt.legend()
plt.title("Multiple Lines")
plt.show()
```

9. Set Axis Limits

```python
plt.plot(x, y)
plt.xlim(0, 6)
plt.ylim(0, 40)
plt.title("Axis Limits")
plt.show()
```

10. Change Line Style and Color

```python
plt.plot(x, y, linestyle='--', color='green', marker='o')
plt.title("Line Style Example")
plt.show()
```

11. Subplots (1 row, 2 columns)

```python
plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title("Plot 1")

plt.subplot(1, 2, 2)
plt.plot(x, z)
plt.title("Plot 2")

plt.tight_layout()
plt.show()
```

12. Save Figure as Image

```python
plt.plot(x, y)
plt.title("Saved Plot")
plt.savefig("my_plot.png")
```

13. Box Plot

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(data)
plt.title("Box Plot")
plt.show()

```

14. Stack Plot

```python
days = [1, 2, 3, 4, 5]
eating = [1, 2, 3, 4, 5]
sleeping = [2, 3, 4, 3, 2]

plt.stackplot(days, eating, sleeping, labels=['Eating', 'Sleeping'])
plt.legend()
plt.title("Stack Plot")
plt.show()
```

15. Fill Between Two Lines

```python
plt.fill_between(x, y, z, color='lightgreen')
plt.title("Fill Between")
plt.show()
```

16. Add Text on Plot

```python
plt.plot(x, y)
plt.text(3, 20, "Mid Point", fontsize=12)
plt.title("Text on Plot")
plt.show()
```

17. Custom Tick Labels

```python
plt.plot(x, y)
plt.xticks([1, 2, 3, 4, 5], ['A', 'B', 'C', 'D', 'E'])
plt.title("Custom Ticks")
plt.show()
```

18. Change Figure 


```python
plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.title("Custom Figure Size")
plt.show()
```

19. Add Annotations

```python
plt.plot(x, y)
plt.annotate("Peak", xy=(5, 30), xytext=(3, 35),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.title("Annotation Example")
plt.show()
```

20. Clear a Plot

```python
plt.plot(x, y)
plt.clf()
```
