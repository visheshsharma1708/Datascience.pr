```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2 * np.pi, 100)

# Calculate sine of x
y = np.sin(x)

# Plot the sine wave
plt.plot(x, y)

# Add labels and title
plt.title('Sine Wave')
plt.xlabel('Angle (radians)')
plt.ylabel('sin(x)')

# Show grid and plot
plt.grid(True)
plt.show()
```
