## Data preprocessing 
 ###  Data cleanning :missing values, outliers, duplicates
    top commands in pandas for cleaning data and doing data preprocessing 

```python
import pandas as pd

csv_1 = pd.read_csv("C:\\Users\\svish\\OneDrive\\Desktop\\titanic-dataset\\test.csv")
csv_1
```

```python
csv_1.isnull().sum()
#check missing values per column 
```

```python
(csv_1.isnull().sum() / len(csv_1)) * 100
# gives percentage of missing values in the data set 
```

```python
# csv_1.dropna(inplace=True)
# # drop rows with having null values 
# csv_1
```

```python
csv_1.dropna(subset=['Age'], inplace=True)
#drop according to specific column
csv_1
```

```python
csv_1['Age'].fillna(csv_1['Age'].mean(), inplace=True)
# Fill missing values in a column with mean
csv_1
```

```python
csv_1['Age'].fillna(csv_1['Age'].mean(), inplace=True)
# Fill missing values in a column with mean
csv_1
```

```python
csv_1['Fare'].fillna(csv_1['Fare'].median(), inplace=True)
#fills missing values wiht the median 
csv_1
```

```python
csv_1['Embarked'].fillna(csv_1['Embarked'].mode()[0], inplace=True)
# fills missing value with the mode 
csv_1
```

```python
csv_1['Cabin'] = csv_1['Cabin'].fillna('Unknown')
csv_1
```

```python
csv_1.fillna(method='ffill')
#forword fill method 
csv_1
```

```python
csv_1.fillna(method='bfill')
#backword filling method 
csv_1
```

outliers 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

csv_1 = pd.read_csv("C:\\Users\\svish\\OneDrive\\Desktop\\titanic-dataset\\test.csv")

sns.boxplot(data=csv_1, x='Fare')
plt.title("Boxplot of Fare Column")
plt.show()
```

```python
Q1 = csv_1['Fare'].quantile(0.25)
Q3 = csv_1['Fare'].quantile(0.75)
IQR = Q3 - Q1
#  Calculate IQR (Interquartile Range) for 'Fare'
```

```python
csv_1 = csv_1[(csv_1['Fare'] >= Q1 - 1.5 * IQR) & (csv_1['Fare'] <= Q3 + 1.5 * IQR)]
#Remove Outliers Based on IQR
csv_1
```

```python
csv_1[['Fare', 'Age']].describe()
```

```python
num_cols = csv_1.select_dtypes(include=['int64', 'float64']).columns
print(num_cols)
```
