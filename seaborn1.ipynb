### Seaborn practice 

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('C:\\Users\\svish\\OneDrive\\Desktop\\titanic-dataset\\train.csv')
```

```python
sns.set()
sns.histplot(data=df, x='Age', kde=True)
```

2.) box plot 

```python
sns.boxplot(data=df, x='Sex', y='Age')
```

3.) violin plot 

```python
sns.violinplot(data=df, x='Sex', y='Age')
```

4.) countplot 

```python
sns.countplot(data=df, x='Survived')
```

5.) barplot 

```python
sns.barplot(data=df, x='Sex', y='Fare')
```

6.) scatter plot 

```python
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived')
```

7.) lineplot 

```python
sns.lineplot(data=df.sort_values('Age'), x='Age', y='Fare')
```

8.) heatmap 

```python
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

9.) pairplot 

```python
sns.pairplot(df[['Age', 'Fare', 'Survived']])
```

10.) catplot 

```python
sns.catplot(data=df, x='Sex', y='Fare', kind='box')
```

11.)swarmplot

```python
sns.swarmplot(data=df, x='Sex', y='Age')
```

12.) stripplot 


```python
sns.stripplot(data=df, x='Sex', y='Age', jitter=True)
```

13.) lmplot

```python
sns.lmplot(data=df, x='Age', y='Fare', hue='Survived')
```

14.) kdeplot

```python
sns.kdeplot(data=df['Age'], shade=True)
```

15.)jointplot

```python
sns.jointplot(data=df, x='Age', y='Fare', kind='scatter')
```

16.) cluster map 

```python
sns.clustermap(df[['Age', 'Fare']].dropna())
```

17.) despline 

```python
sns.despine()
#Removes top and right spines (borders). Makes the plot cleaner.
```

18.) color palette

```python
sns.set_palette("pastel")
```

19.) facetgrid 

```python
g = sns.FacetGrid(df, col="Sex")
g.map(sns.histplot, "Age")
```
