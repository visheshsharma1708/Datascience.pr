```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
import pandas as pd

# Data dictionary
data = {
    "area": [800, 1000, 1200, 1500, 1800, 2000, 2200, 2500, 2800, 3000],  # in square feet
    "price": [40, 55, 68, 85, 100, 112, 125, 140, 160, 175]  # in lakhs (â‚¹)
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Save to CSV
df.to_csv("homeprice.csv", index=False)

print("CSV file 'homeprice.csv' created successfully.")
df
```

```python

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load data
df = pd.read_csv("homeprice.csv")

# Train model
reg = LinearRegression()
reg.fit(df[["area"]], df["price"])

# Plot scatter
plt.scatter(df["area"], df["price"], color='blue', marker='o')
plt.xlabel("Area (sq. ft.)")
plt.ylabel("Price (in â‚¹ Lakh)")
plt.title("Area vs Price of Homes")

# Plot regression line
plt.plot(df["area"], reg.predict(df[["area"]]), color='red', label="Regression Line")

# Add legend and show plot
plt.legend()
plt.grid(True)
plt.show()
```

```python
reg= linear_model.LinearRegression()
reg.fit(df[["area"]],df["price"])
```

```python
reg.predict([[2200]])
```

```python
print("Slope (m):", reg.coef_[0])
print("Intercept (c):", reg.intercept_)
```
