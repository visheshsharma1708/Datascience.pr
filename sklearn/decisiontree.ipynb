## Decision tree 
comapany salary and degree example 

```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier

# Step 2: Create a larger dataset
data = {
    'Company': ['Google', 'TCS', 'Infosys', 'Google', 'TCS', 'Infosys', 'Google', 'Infosys', 'TCS', 'Google',
                'TCS', 'Infosys', 'Google', 'TCS', 'Infosys'],
    'Degree': ['Masters', 'Bachelors', 'Bachelors', 'Masters', 'Masters', 'Bachelors', 'Bachelors',
               'Masters', 'Bachelors', 'Masters', 'Masters', 'Masters', 'Bachelors', 'Bachelors', 'Masters'],
    'Salary >500K': ['Yes', 'No', 'No', 'Yes', 'No', 'No', 'Yes', 'No', 'No', 'Yes', 'No', 'Yes', 'Yes', 'No', 'No']
}

df = pd.DataFrame(data)

# Step 3: Map categories to numbers
company_map = {'Google': 0, 'TCS': 1, 'Infosys': 2}
degree_map = {'Bachelors': 0, 'Masters': 1}
salary_map = {'No': 0, 'Yes': 1}

df['Company'] = df['Company'].map(company_map)
df['Degree'] = df['Degree'].map(degree_map)
df['Salary >500K'] = df['Salary >500K'].map(salary_map)

# Step 4: Features and target
X = df[['Company', 'Degree']]
y = df['Salary >500K']

# Step 5: Train Decision Tree model
model = DecisionTreeClassifier()
model.fit(X, y)

# Step 6: Scatter plot (Company vs Degree)
plt.figure(figsize=(8, 5))
colors = ['red' if val == 0 else 'green' for val in df['Salary >500K']]
plt.scatter(df['Company'], df['Degree'], c=colors)
plt.xlabel("Company (0=Google, 1=TCS, 2=Infosys)")
plt.ylabel("Degree (0=Bachelors, 1=Masters)")
plt.title("Employee Salary > 500K (Green = Yes, Red = No)")
plt.grid(True)
plt.show()

# Step 7: Make multiple predictions
test_inputs = [
    [0, 0],  # Google + Bachelors
    [0, 1],  # Google + Masters
    [1, 0],  # TCS + Bachelors
    [2, 1],  # Infosys + Masters
    [2, 0],  # Infosys + Bachelors
]

for company, degree in test_inputs:
    result = model.predict([[company, degree]])[0]
    status = "Yes" if result == 1 else "No"
    print(f"Company: {company}, Degree: {degree} → Salary > 500K? {status}")
```
