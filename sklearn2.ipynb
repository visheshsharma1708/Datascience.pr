
## Imputation methods 
### 1.) simple imputer 

```python
import pandas as pd
import seaborn as sns
from sklearn.impute import SimpleImputer

df = pd.read_csv("C:\\Users\\svish\\OneDrive\\Desktop\\titanic-dataset\\train.csv")
age_data = df[["Age"]]  

# Create imputer
imputer = SimpleImputer(strategy='mean')  

# Apply imputer b
age_imputed = imputer.fit_transform(age_data)

print("Before:", age_data.isnull().sum())
print("After Imputation:", pd.DataFrame(age_imputed).isnull().sum())
df
```

### 2.) KNN imputer 

```python
from sklearn.impute import KNNImputer

numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
knn_imputer = KNNImputer(n_neighbors=3)

knn_result = knn_imputer.fit_transform(df[numerical_cols])
print("Imputation done using KNN")
df
```

### 3.) iterartive imputer (mice)


```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

mice_imputer = IterativeImputer()
mice_result = mice_imputer.fit_transform(df[numerical_cols])
print("Imputation done using MICE (IterativeImputer)")
df
```

### scaler methods 

1.) standard scaler- z score normalization

```python

from sklearn.preprocessing import StandardScaler

df = pd.read_csv("C:\\Users\\svish\\OneDrive\\Desktop\\titanic-dataset\\train.csv")

df_numeric = df[['Age', 'Fare']].dropna()


scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_numeric)
print("Scaled Age Mean:", scaled_data[:, 0].mean())
print("Scaled Fare Mean:", scaled_data[:, 1].mean())
```

2.) minmax scaler

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
minmax_scaled = scaler.fit_transform(df_numeric)

print("Age Range after MinMaxScaler:", minmax_scaled[:, 0].min(), "to", minmax_scaled[:, 0].max())
```

3.) normalizer

```python
import pandas as pd
from sklearn.preprocessing import Normalizer

df = pd.read_csv("C:\\Users\\svish\\OneDrive\\Desktop\\titanic-dataset\\train.csv")

df_numeric = df[['Age', 'Fare']].dropna()

normalizer = Normalizer()
normalized_data = normalizer.fit_transform(df_numeric)

print("First 5 normalized rows:")
print(normalized_data[:5])
```

4.) PowerTransformer (Box-Cox or Yeo-Johnson)

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import PowerTransformer

power = PowerTransformer(method='yeo-johnson')
transformed_data = power.fit_transform(df_numeric)

print("Skewness Removed using Yeo-Johnson")

# Before transformation
sns.histplot(df_numeric['Fare'], kde=True)
plt.title("Before Yeo-Johnson (Fare)")
plt.show()

# After transformation
sns.histplot(transformed_data[:, 1], kde=True)
plt.title("After Yeo-Johnson (Fare)")
plt.show()
```

5.) PolynomialFeatures

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2, include_bias=False)
poly_features = poly.fit_transform(df_numeric)

print("Shape before:", df_numeric.shape)
print("Shape after PolynomialFeatures:", poly_features.shape)
```

### Handling categorical variables 
1 LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder

df = sns.load_dataset('titanic')
le = LabelEncoder()

df_temp = df.copy()
df_temp['sex_encoded'] = le.fit_transform(df_temp['sex'])

df_temp[['sex', 'sex_encoded']].head()
```

2 OneHotEncoder

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Load dataset
df = pd.read_csv("C:\\Users\\svish\\OneDrive\\Desktop\\titanic-dataset\\train.csv")

# Create OneHotEncoder object
encoder = OneHotEncoder(sparse_output=False, drop='first')  # drop='first' to avoid dummy variable trap

# Handle missing values in 'Embarked'
embarked_data = df[['Embarked']].fillna('missing')

# Fit and transform
encoded = encoder.fit_transform(embarked_data)

# Create new DataFrame with encoded columns
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(['Embarked']))

# Combine original df and new encoded_df
df_ohe = pd.concat([df, encoded_df], axis=1)

# Show result
print(encoded_df.head())
```

outiler detection and visualization

```python

df = df.dropna(subset=['Fare'])

# Step 1: Calculate Q1, Q3, and IQR
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1

# Step 2: Find Outliers
outliers = df[(df['Fare'] < (Q1 - 1.5 * IQR)) | (df['Fare'] > (Q3 + 1.5 * IQR))]

# Step 3: Print how many outliers found
print("Number of outliers in 'Fare':", outliers.shape[0])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 4))
sns.boxplot(x=df['Fare'])
plt.title("Boxplot for Fare")
plt.show()
```
